```python
import requests as re
from bs4 import BeautifulSoup
def extracturl(url):
    req = re.get(url)
    soup = BeautifulSoup(req.content,"html5lib")
    
    return soup.title.text

url=["https://www.geeksforgeeks.org/taking-input-in-python/?ref=outind","https://www.geeksforgeeks.org/"]
for i in range(0,2):
     print(i,":",extracturl(url[i]))
```
